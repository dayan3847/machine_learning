```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from dayan3847.models.Model import Model
from dayan3847.models.ModelTrainer import ModelTrainer
from dayan3847.models.linear.LinearPolynomialModel import LinearPolynomialModel
from dayan3847.models.linear.LinearGaussianModel import LinearGaussianModel
from dayan3847.models.linear.LinearSigmoidalModel import LinearSigmoidalModel
```

```python
np.random.seed(0)
```

```python
data: np.array = np.loadtxt('data_cos_2_pi_x.csv', delimiter=',').T
```

```python
def plot_data_cos_2_pi_x():
    global data
    _x: np.array = np.linspace(0, 1, 100)
    _y: np.array = np.cos(2 * np.pi * _x)
    plt.plot(_x, _y, label='cos(2πx)', color='gray', linestyle='--')
    plt.scatter(data[0], data[1], label='data', color='gray', marker='o')
    plt.xlabel('x')
    plt.ylabel('y')


plot_data_cos_2_pi_x()
plt.legend()
plt.show()
```

```python
epochs_count: int = 100
```

## Polynomial Model

```python
linear_polynomial_model: Model = LinearPolynomialModel(
    a=.1,
    f=5,
)
```

```python
def plot_linear_polynomial_model():
    global linear_polynomial_model
    _x: np.array = np.linspace(0, 1, 100)
    _y: np.array = linear_polynomial_model.g(_x)
    plt.plot(_x, _y, label='polynomial', c='b', linestyle='-')


plot_data_cos_2_pi_x()
plot_linear_polynomial_model()
plt.legend()
plt.show()
```

```python
trainer_polynomial: ModelTrainer = ModelTrainer(
    linear_polynomial_model,
    epochs_count,
    data[0],
    data[1],
)
trainer_polynomial.train()
```

```python
plot_data_cos_2_pi_x()
plot_linear_polynomial_model()
plt.legend()
plt.show()
```

```python
plt.title('Error')
plt.plot(trainer_polynomial.error_history, label='polynomial', c='r', linestyle='-')
plt.legend()
plt.show()
```

## Gaussian Model

```python
linear_gaussian_model = LinearGaussianModel(
    a=.1,
    f=5,
    mu_lim=(0, 1),
    s=.1
)
```

```python
def plot_linear_gaussian_model():
    global linear_gaussian_model
    _x: np.array = np.linspace(0, 1, 100)
    _y: np.array = linear_gaussian_model.g(_x)
    plt.plot(_x, _y, label='gaussian', c='g', linestyle='-')


plot_data_cos_2_pi_x()
plot_linear_gaussian_model()
plt.legend()
plt.show()
```

```python
trainer_gaussian: ModelTrainer = ModelTrainer(
    linear_gaussian_model,
    epochs_count,
    data[0],
    data[1],
)
trainer_gaussian.train()
```

```python
plot_data_cos_2_pi_x()
plot_linear_gaussian_model()
plt.legend()
plt.show()
```

```python
plt.title('Error')
plt.plot(trainer_gaussian.error_history, label='gaussian', c='r', linestyle='-')
plt.legend()
plt.show()
```

```python
linear_sigmoidal_model = LinearSigmoidalModel(
    a=.1,
    f=5,
    mu_lim=(0, 1),
    s=.1
)
```
