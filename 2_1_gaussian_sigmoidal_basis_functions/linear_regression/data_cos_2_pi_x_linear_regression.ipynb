```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from dayan3847.models.Model import Model
from dayan3847.models.linear.LinearPolynomialModel import LinearPolynomialModel
from dayan3847.models.linear.LinearGaussianModel import LinearGaussianModel
from dayan3847.models.linear.LinearSigmoidalModel import LinearSigmoidalModel
from dayan3847.models.functions import train_model
```

```python
np.random.seed(0)
```

```python
data: np.array = np.loadtxt('data_cos_2_pi_x.csv', delimiter=',')
data.shape
```

```python
def plot_data_cos_2_pi_x():
    global data
    _x: np.array = np.linspace(0, 1, 100)
    _y: np.array = np.cos(2 * np.pi * _x)
    plt.plot(_x, _y, label='cos(2πx)', color='gray', linestyle='--')
    plt.scatter(data[:, 0], data[:, 1], label='data', color='gray', marker='o')
    plt.xlabel('x')
    plt.ylabel('y')


plot_data_cos_2_pi_x()
plt.legend()
plt.show()
```

## Polynomial Model

```python
linear_polynomial_model: Model = LinearPolynomialModel(5)
```

```python
def plot_linear_polynomial_model():
    global linear_polynomial_model
    _x: np.array = np.linspace(0, 1, 100)
    _y: np.array = linear_polynomial_model.g(_x)
    plt.plot(_x, _y, label='polynomial', c='b', linestyle='-')


plot_data_cos_2_pi_x()
plot_linear_polynomial_model()
plt.legend()
plt.show()
```

```python
error_history_polynomial = train_model(
    linear_polynomial_model,
    data[:, 0],
    data[:, 1],
    a=.1,
    epochs_count=100,
    error_threshold=2,
)
```

```python
plot_data_cos_2_pi_x()
plot_linear_polynomial_model()
plt.legend()
plt.show()
```

```python
plt.title('Error')
plt.plot(error_history_polynomial, label='polynomial', c='r', linestyle='-')
plt.legend()
plt.show()
```

```python
plot_data_cos_2_pi_x()
plot_linear_gaussian_model()
plt.legend()
plt.show()
```

```python
plt.title('Error')
plt.plot(error_history_gaussian, label='gaussian', c='r', linestyle='-')
plt.legend()
plt.show()
```
