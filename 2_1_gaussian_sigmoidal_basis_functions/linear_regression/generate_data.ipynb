```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
points_number: int = 100
```

## sin(2πx)

```python
sp.sin(2 * sp.pi * sp.Symbol('x'))
```

```python
def generate_data_sin_2_pi_x() -> np.array:
    _x: np.array = np.random.rand(points_number)
    _e: np.array = np.random.rand(points_number) * .6 - .3
    _y: np.array = np.sin(2 * np.pi * _x) + _e
    return np.array([_x, _y])


data_sin_2_pi_x = generate_data_sin_2_pi_x()
```

```python
def plot_data_sin_2_pi_x():
    global data_sin_2_pi_x
    _x: np.array = np.linspace(0, 1, 100)
    _y: np.array = np.sin(2 * np.pi * _x)
    plt.plot(_x, _y, label='sin(2πx)', color='gray', linestyle='--')
    plt.scatter(data_sin_2_pi_x[0], data_sin_2_pi_x[1], label='data', color='gray', marker='o')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()


plot_data_sin_2_pi_x()
```

```python
np.savetxt('data_sin_2_pi_x.csv', data_sin_2_pi_x.T, delimiter=',')
```
