## Polynomial

```python
# Count of factors
count: int = 3
'count: {}'.format(count)
```

```python
def factor_polynomial() -> sp.Symbol:
    # Variable
    x: sp.Symbol = sp.Symbol('x')
    # Hyper-Parameters
    n: sp.Symbol = sp.Symbol('n')  # Order of factor (degree)
    return x ** n


factor_polynomial()
```

```python
# Degree list
n_list: np.array = np.arange(count)
n_list
```

```python
def equation_polynomial() -> sp.Symbol:
    result: sp.Symbol = 0
    # Variable
    x = sp.Symbol('x')
    for i, n in zip(np.arange(count), n_list):
        wi: sp.Symbol = sp.symbols('w{}'.format(i))
        result += wi * (x ** n)

    return result


equation_polynomial()
```

```python
# Plot
def plot_polynomial(w_list_):
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
    x_plot: np.array = np.linspace(-1, 1, 1000)

    y_plot_e: np.array = np.full(len(x_plot), .0)
    for w, n in zip(w_list_, n_list):
        y_plot_f: np.array = w * (x_plot ** n)
        ax[0].plot(x_plot, y_plot_f)
        y_plot_e = y_plot_e + y_plot_f

    ax[0].set_title('Polynomial Factors')
    ax[1].set_title('Polynomial Equation')
    ax[1].plot(x_plot, y_plot_e)

    for i in range(2):
        ax[i].grid(True)
        ax[i].set_xlabel('x')
        ax[i].set_ylabel('y')

    plt.tight_layout()
    plt.show()
```

### Weight = 1

```python
w_list = np.full(count, 1)
plot_polynomial(w_list)
```

### Weight = [1] 5 [1]

```python
w_list[int(count / 2)] = 5
plot_polynomial(w_list)
```

```python
w_list = np.random.rand(count) - .5
plot_polynomial(w_list)
```

## Gaussian

```python
def factor_gaussian() -> sp.exp:
    # Variable
    x: sp.Symbol = sp.Symbol('x')
    # Hyper-Parameters
    m: sp.Symbol = sp.Symbol('m')  # Mean
    s: sp.Symbol = sp.Symbol('s')  # Standard deviation
    return sp.exp(-1 * ((x - m) ** 2) / (2 * (s ** 2)))


factor_gaussian()
```

```python
config = 0, 2, 5
```

```python
# Factors Count
'Factors Count: {}'.format(config[2])
```

```python
f_step: float = (config[1] - config[0]) / config[2]
'Factor Step: {}'.format(f_step)
```

```python
# Mean list
m_list: np.array = np.linspace(*config)
m_list
```

```python
# Standard deviation list
s2: float = 5e-2 * f_step
s2
```

```python
# Plot
def plot_gaussian(w_list_):
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
    x_plot: np.array = np.linspace(config[0], config[1], 1000)

    y_plot_e: np.array = np.full(len(x_plot), .0)
    for w, m in zip(w_list_, m_list):
        y_plot_f: np.array = w * np.exp(-.5 * ((x_plot - m) ** 2) / s2)
        ax[0].plot(x_plot, y_plot_f)
        y_plot_e = y_plot_e + y_plot_f

    ax[0].set_title('Gaussian Factors')
    ax[1].set_title('Gaussian Equation')
    ax[1].plot(x_plot, y_plot_e)

    for i in range(2):
        ax[i].grid(True)
        ax[i].set_xlabel('x')
        ax[i].set_ylabel('y')

    plt.tight_layout()
    plt.show()
```

```python
w_list = np.full(count, 1)
plot_gaussian(w_list)
```

```python
w_list[2] = -2
# w_list[3] = 2
plot_gaussian(w_list)
```

```python
w_list = np.random.rand(count) - .5
plot_gaussian(w_list)
```

## Sigmoidal

```python
# Count of factors
count: int = 3
'count: {}'.format(count)
```

```python
def factor_sigmoidal() -> sp.Expr:
    # Variable
    x: sp.Symbol = sp.Symbol('x')
    # Hyper-Parameters
    m: sp.Symbol = sp.Symbol('m')  # Mean
    s: sp.Symbol = sp.Symbol('s')  # Standard deviation
    return 1 / (1 + sp.exp((m - x) / s))


factor_sigmoidal()
```

```python
# Mean list
m_list: np.array = np.linspace(-1, 1, count)
m_list
```

```python
# Standard deviation list
s_list: np.array = np.full(count, 0.1)
s_list
```

```python
def plot_sigmoidal(w_list_):
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
    x_plot: np.array = np.linspace(-1, 1, 1000)

    y_plot_e: np.array = np.full(len(x_plot), .0)
    for w, m, s in zip(w_list_, m_list, s_list):
        y_plot_f: np.array = w * (1 / (1 + np.exp((m - x_plot) / s)))
        ax[0].plot(x_plot, y_plot_f)
        y_plot_e = y_plot_e + y_plot_f

    ax[0].set_title('Sigmoidal Factors')
    ax[1].set_title('Sigmoidal Equation')
    ax[1].plot(x_plot, y_plot_e)

    for i in range(2):
        ax[i].grid(True)
        ax[i].set_xlabel('x')
        ax[i].set_ylabel('y')

    plt.tight_layout()
    plt.show()
```

```python
w_list = np.full(count, 1)
plot_sigmoidal(w_list)
```

```python
w_list[int(count / 2)] = 2
plot_sigmoidal(w_list)
```

```python
w_list = np.random.rand(count) - .5
plot_sigmoidal(w_list)
```
