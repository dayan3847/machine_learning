### Imports

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import plotly as py
```

### Random Seed

```python
np.random.seed(0)
```

## Gaussian

```python
def factor_gaussian() -> sp.exp:
    # Variable
    x, y = sp.symbols('x y')
    # Hyper-Parameters
    m1, m2 = sp.symbols('m1 m2')  # Mean
    s: sp.Symbol = sp.Symbol('s')  # Standard deviation
    return sp.exp(-1 * ((x - m1) ** 2) / (2 * (s ** 2))) * sp.exp(-1 * ((x - m2) ** 2) / (2 * (s ** 2)))


factor_gaussian()
```

```python
def equation_gaussian() -> sp.Symbol:
    # Variable
    w, x, y = sp.symbols('w x y')
    return w * ( sp.exp(-1 * (x ** 2) / 2) ) * ( sp.exp(-1 * (y ** 2) / 2) )

equation_gaussian()
```

```python
# Plot
def plot_gaussian(w):
    S = 3
    fig = plt.figure(figsize=(10, 5))
    ax = [
        fig.add_subplot(121, projection='3d'),
        fig.add_subplot(122, projection='3d'),
    ]

    for i in range(2):
        ax[i].set_xlim(-S, S)
        ax[i].set_ylim(-S, S)
        ax[i].set_zlim(0, 1)
        ax[i].set_xlabel('x')
        ax[i].set_ylabel('y')
        ax[i].set_zlabel('z')

    ax[0].set_title('Gaussian without Activation Function')
    ax[1].set_title('Gaussian Equation with Activation Function (sigmoid)')

    x_plot, y_plot = np.meshgrid(
        np.linspace(-S, S, 100),
        np.linspace(-S, S, 100)
    )

    z_plot_g = w * (np.exp(-1 * (x_plot ** 2) / 2)) * (np.exp(-1 * (y_plot ** 2) / 2)) + 0.5
    z_plot_gs = 1 / (1 + np.exp(-1 * z_plot_g))

    ax[0].plot_surface(x_plot, y_plot, z_plot_g, cmap='viridis')

    ax[1].plot_surface(x_plot, y_plot, z_plot_gs, cmap='viridis')

    plt.tight_layout()
    plt.show()
```

```python
plot_gaussian(.5)
```

```python
plot_gaussian(5)
```

```python
plot_gaussian(-5)
```

```python
# Plot (with Plotly)
def plot_gaussian_py(w):
    S = 3
    x_plot, y_plot = np.meshgrid(
        np.linspace(-S, S, 100),
        np.linspace(-S, S, 100)
    )

    z_plot_g = w * (np.exp(-1 * (x_plot ** 2) / 2)) * (np.exp(-1 * (y_plot ** 2) / 2)) + 0.5
    z_plot_gs = 1 / (1 + np.exp(-1 * z_plot_g))

    fig = py.graph_objs.Figure(data=[py.graph_objs.Surface(z=z_plot_g)])
    fig.show()

    fig = py.graph_objs.Figure(data=[py.graph_objs.Surface(z=z_plot_gs)])
    fig.show()
```

```python
plot_gaussian_py(.5)
```

```python
plot_gaussian_py(5)
```

```python
plot_gaussian_py(-5)
```
