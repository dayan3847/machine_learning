```python
import numpy as np
from tensorflow import keras
```

```python
img_path = 'elefante.jpg'
img_path = 'elefante_india.jpg'
```

```python
img = keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
```

```python
# see the image
img
```

```python
img.size
```

```python
# img_to_array
img_array = keras.preprocessing.image.img_to_array(img)
img_array.shape
```

```python
# save 
np.save('elefante.npy', img_array)
```
