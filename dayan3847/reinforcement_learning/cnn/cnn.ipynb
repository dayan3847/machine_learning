```python
import numpy as np
from tensorflow import keras
```

```python
model_resnet50 = keras.applications.resnet50.ResNet50(include_top=False, weights='imagenet', input_shape=(240, 320, 3))
model_resnet50.summary()
```

```python
image_array = np.load('elefante.npy')
image_array.shape
```

```python
images = np.array([
    image_array,
])
images.shape
```

```python
p = model_resnet50.predict(images)
p.shape
```
