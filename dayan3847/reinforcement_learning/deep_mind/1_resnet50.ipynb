```python
import numpy as np
from tensorflow import keras
```

```python
model_resnet50 = keras.applications.resnet50.ResNet50(
    include_top=False,
    weights='imagenet',
    input_shape=(240, 320, 3),
)
```

```python
model_resnet50.summary()
```

```python
# load frame.jpg
img = keras.preprocessing.image.load_img(
    "frame.jpg", 
    # target_size=(240, 320),
)
```

```python
x = keras.preprocessing.image.img_to_array(img)
x.shape
```

```python
intput = np.array([x])
intput.shape
```

```python
output = model_resnet50.predict(intput)
output.shape
```
