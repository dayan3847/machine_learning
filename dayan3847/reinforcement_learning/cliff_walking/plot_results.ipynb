```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.title('Reward')
plt.plot(np.loadtxt('qlearning_reward.txt', delimiter=','), label='qlearning')
plt.plot(np.loadtxt('sarsa_reward.txt', delimiter=','), label='sarsa')
plt.legend()
```
