```python
import numpy as np
import matplotlib.pyplot as plt
from Agent import Environment, AgentQLearning, AgentQLearningTable
from Trainner import Trainner
```

```python
np.random.seed(0)
```

```python
env: Environment = Environment()
agent: AgentQLearningTable = AgentQLearningTable(env)
trainer = Trainner(agent, 20, 500)
```

```python
trainer.train_callback()
```

```python
trainer.agent.save_knowledge()
```

```python
# Plot the Reward Average
plt.plot(trainer.get_rewards_average(), label='Reward Average')
plt.xlabel('Episode')
plt.ylabel('Reward Average')
plt.show()
```

```python
# Plot the Success
plt.plot(trainer.success, label='Success')
plt.xlabel('Episode')
plt.ylabel('Success')
plt.show()
```

```python
# Plot the Best Action
plt.matshow(trainer.agent.get_best_actions_for_all_states(), label='Best Action')
plt.xlabel('State _0')
plt.ylabel('State _1')
plt.show()
```

```python
# Plot the Q-Table (Action: UP)
# matshow
table_up = agent.q_tables[0]
plt.matshow(table_up)
plt.title('Q-Table (Action: UP)')
plt.xlabel('State _0')
plt.ylabel('State _1')
plt.show()
```
